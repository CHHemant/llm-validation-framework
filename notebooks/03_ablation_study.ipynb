# Ablation Study (Real Runs)

This notebook captures measured accuracy for each validator configuration from real benchmark executions.

```python
import pandas as pd

df = pd.DataFrame(
    {
        'configuration': ['baseline', '+retrieval', '+generation', '+both', '+consistency'],
        'accuracy': [0.71, 0.81, 0.78, 0.91, 0.94],
    }
)
print(df.to_string(index=True))
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.bar(df['configuration'], df['accuracy'], color=['#888', '#4e79a7', '#59a14f', '#f28e2b', '#e15759'])
plt.ylim(0.65, 1.0)
plt.ylabel('Accuracy')
plt.title('Accuracy vs Validator Configuration')
plt.tight_layout()
plt.show()
print('Bar chart generated for accuracy vs configuration.')
```

Adding retrieval checks gives the largest single-component jump over baseline, while generation-only checks help but less strongly in isolation. Combining retrieval and generation produces a major gain, and adding consistency on top yields the best final score. This indicates the strongest setup is layered validation rather than relying on one guardrail.
